    To simulate the relationship between a dependent variable and one or more independent variables, a statistical method called a regression model is used. Regression analysis seeks to identify the line that most accurately depicts the relationship between the dependent and independent variables. Based on the new values for the independent variables, predictions about the dependent variable can then be made using this model. Regression models can be applied to a number of real-world issues, including forecasting future sales, determining a home's value based on its features, and understanding the relationship between a person's height and weight. Regression models that are most frequently used include logistic, polynomial, and linear regression. The appropriate regression model depends on the particular problem at hand, the nature of the dependent and independent variables, and the availability of data. In many instances, linear regression is a good place to start, but more complex models might be needed if there are outliers in the data or if the relationship between the variables is not linear.

    Linear Regression - The statistical technique of linear regression involves fitting a linear equation to the observed data in order to model the relationship between a dependent variable and one or more independent variables. One of the most well-liked machine learning libraries for Python is the scikit-learn library, and it can be used to perform linear regression.

```python
# Step 1: INSTANTIATE a model object
lm_best = smf.ols(formula =  """REVENUE ~ AVG_TIME_PER_SITE_VISIT +
                                          TOTAL_MEALS_ORDERED +
                                          TOTAL_PHOTOS_VIEWED + 
                                          LARGEST_ORDER_SIZE +
                                          CONTACTS_W_CUSTOMER_SERVICE +
                                          AVG_PREP_VID_TIME +
                                          UNIQUE_MEALS_PURCH +
                                          PRODUCT_CATEGORIES_VIEWED +
                                          CANCELLATIONS_AFTER_NOON +
                                          PC_LOGINS +
                                          MOBILE_LOGINS +
                                          WEEKLY_PLAN +
                                          AVG_MEAN_RATING""",
                                data = cd)

# Step 2: FIT the data into the model object
results = lm_best.fit()


# Step 3: analyze the SUMMARY output
print(results.summary())
```

```python
# x-variables for normal y
x_var = ["AVG_TIME_PER_SITE_VISIT",
        "AVG_PREP_VID_TIME",
        "AVG_MEAN_RATING",
        "TOTAL_PHOTOS_VIEWED",
        "TOTAL_MEALS_ORDERED",
        "UNIQUE_MEALS_PURCH",
        "CONTACTS_W_CUSTOMER_SERVICE",
        "PRODUCT_CATEGORIES_VIEWED",
        "CANCELLATIONS_AFTER_NOON",
        "PC_LOGINS",
        "MOBILE_LOGINS",
        "WEEKLY_PLAN",
        "LARGEST_ORDER_SIZE",
        "LATE_DELIVERIES "]
```

```python
# preparing for scikit-learn

# Preparing a DataFrame based the the analysis above
x_data = cd[x_var]


# preparing response variable
y_data = np.log(cd["REVENUE"])


#################################
## setting up train-test split ##
#################################

x_train, x_test, y_train, y_test = train_test_split(
            x_data, 
            y_data,
            test_size    = 0.25,
            random_state = 219)
```

LINEAR REGRESSION

```python
# Setting a model name
model_name = "Linear Regression"


# INSTANTIATING a model object - CHANGE THIS AS NEEDED
model = LinearRegression()


# FITTING to the training data
model_fit = model.fit(x_train, y_train)


# PREDICTING on new data
test_pred = model.predict(x_test)


# SCORING the results
print('OLS Training Score :', model.score(x_train, y_train).round(4))
print('OLS Testing Score  :', model.score(x_test, y_test).round(4))


# saving scoring data for future use
lr_train_score = model.score(x_train, y_train).round(4) # using R-square
lr_test_score  = model.score(x_test, y_test).round(4)   # using R-square


# displaying and saving the gap between training and testing
print('OLS Train-Test Gap :', abs(lr_train_score - lr_test_score).round(4))
lr_test_gap = abs(lr_train_score - lr_test_score).round(4)
```

LASSO REGRESSION

    LASSO REGRESSION - The linear regression technique known as "lasso regression" employs shrinkage. The loss function is modified by a penalty term whose absolute value is proportional to the coefficients' absolute values. In order to lower overfitting and improve the model's interpretability, the regularization term aims to bring the coefficients to zero. (Smith, 2020)

```python
import sklearn

# Setting a model name
model_name = "Lasso Regression"


# INSTANTIATING a model object - CHANGE THIS AS NEEDED
model = sklearn.linear_model.Lasso()


# FITTING to the training data
model_fit = model.fit(x_train, y_train)


# PREDICTING on new data
model_pred = model.predict(x_test)


# SCORING the results
l_train_score = model.score(x_train, y_train).round(4) # using R-square
l_test_score  = model.score(x_test, y_test).round(4)   # using R-square


# displaying and saving the gap between training and testing
print('Lasso Training Score :', l_train_score)
print('Lasso Testing Score :',l_test_score)
print('Lasso Train-Test Gap :', abs(l_train_score - l_test_score).round(4))
l_test_gap = abs(l_train_score - l_test_score).round(4)
```

    ARD Regression - Automatic Relevance Determination (ARD) regression, also known as Bayesian linear regression, is a type of linear regression that makes use of a prior distribution over the parameters to improve the model's ability to determine the applicability of the input features. The specific hyperparameter for each feature in an ARD regression determines how strongly the feature is related to the target variable.

```python
# Setting a model name
model_name = "ARD Regression"


# INSTANTIATING a model object - CHANGE THIS AS NEEDED
model = sklearn.linear_model.ARDRegression()


# FITTING to the training data
model_fit = model.fit(x_train, y_train)


# PREDICTING on new data
model_pred = model.predict(x_test)


# SCORING the results
ard_train_score = model.score(x_train, y_train).round(4) # using R-square
ard_test_score  = model.score(x_test, y_test).round(4)   # using R-square
ard_test_gap         = abs(ard_train_score - ard_test_score).round(4)

# displaying and saving the gap between training and testing
print('ARD Training Score :', ard_train_score)
print('ARD Testing Score :',ard_test_score)
print('ARD Train-Test Gap :', abs(ard_train_score - ard_test_score).round(4))
```

GRADIENT BOOSTING REGRESSOR

    A popular machine learning algorithm for classification and regression issues is gradient boosting. By combining a number of weak learners, this ensemble technique creates a strong model.

```python
from sklearn.ensemble import GradientBoostingRegressor # gbm

# specifying a model name
gbd_model_name = 'Gradient Boosting Regressor'


# INSTANTIATING the model object without hyperparameters
model = GradientBoostingRegressor(n_estimators  = 100,
                                  learning_rate = 0.1,
                                  loss          = 'squared_error',
                                  max_depth     = 1,
                                  random_state = 219)


# FITTING the training data
model_fit = model.fit(x_train, y_train)


# PREDICTING based on the testing set
model_pred = model.predict(x_test)


# SCORING the results
gbd_train_score = model.score(x_train, y_train).round(4) # using R-square
gbd_test_score  = model.score(x_test, y_test).round(4)   # using R-square
gbd_test_gap         = abs(gbd_train_score - gbd_test_score).round(4)


# displaying results
print('Training Score :', gbd_train_score)
print('Testing Score  :', gbd_test_score)
print('Train-Test Gap :', gbd_test_gap)
```
